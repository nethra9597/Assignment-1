```python
#IMPORTS
import matplotlib.pyplot as plt
import pandas as pd
import datetime
import numpy as np
import seaborn as sns
Employment_initial_dataframe = pd.read_csv("Employment - National - Weekly (1).csv")
employment_data = Employment_initial_dataframe
```

```python
#Converting to datetime format for convenient plotting
employment_data['Date'] = employment_data['month'].astype(str).str.zfill(2)  + employment_data['day_endofweek'].astype(str).str.zfill(2) + employment_data['year'].astype(str)
employment_data['Date'] = pd.to_datetime(employment_data['Date'], format='%m%d%Y')
employment_data
```

```python
#Duplicating data into a new variable for consistency and backup
plot_data=employment_data

#Setting the date as the index for quarterly plotting
plot_data.set_index('Date')
plot_data
```

```python
#Code block to plot the Income Quartiles in a single graph
sns.set(rc = {'figure.figsize':(40,10)})
sns.set_theme(style="darkgrid")

sns.lineplot(data = plot_data, x='Date' , y='emp_incq1', label ='Quartile 1')
sns.lineplot(data = plot_data, x='Date' , y='emp_incq2', label = 'Quartile 2')
sns.lineplot(data = plot_data, x='Date' , y='emp_incq3', label ='Quartile 3')
sns.lineplot(data = plot_data, x='Date' , y='emp_incq4', label = 'Quartile 4')

plt.xticks(rotation = 45)
plt.title('Employment % change based on Income Quartiles',fontdict={'fontsize': 25})
plt.xlabel('Date', fontdict={'fontsize': 20})
plt.ylabel('Income Quartile', fontdict={'fontsize': 20})
plt.show()
```

```python
#Code block to plot the Employment change based on Type of Industries in a single graph.
sns.set(rc = {'figure.figsize':(40,10)})
sns.set_theme(style="darkgrid")

sns.lineplot(data = plot_data, x='Date' , y='emp_s72',color='brown', label ='Accommodation and Food Services')
sns.lineplot(data = plot_data, x='Date' , y='emp_ss40',color='red', label = 'Trade, Transportation and Utilities ')
sns.lineplot(data = plot_data, x='Date' , y='emp_ss60', label ='Professional and Business Services')
sns.lineplot(data = plot_data, x='Date' , y='emp_ss65',color='black', label = 'Education and Health Services')
sns.lineplot(data = plot_data, x='Date' , y='emp_ss70',color='purple', label = 'Leisure and Hospitality')

plt.xticks(rotation = 45)
plt.title('Employment % change based on Type of Industry', fontdict={'fontsize': 25})
plt.xlabel('Date',fontdict={'fontsize': 20})
plt.ylabel('% change of Types of Industries', fontdict={'fontsize': 20})
plt.show()
```
